```python
#------------------------------------------#

#Metodo di raggruppamento .groupby()

#Analizziamo che categorie ci sono nella variabile 'drive-wheels'
df['drive-wheels'].unique()
```

```python
#Creiamo un nuovo dataframe composto dalle variabili 'drive-wheels', 'body-style' e 'price'

df_group_one=df[['drive-wheels', 'body-style', 'price']]
```

```python
#Raggruppiamo il df con .groupby() e calcoliamo il prezzo medio di ogni categoria

df_group_one = df_group_one.groupby(['drive-wheels'], as_index=False).mean()
df_group_one
```

```python
#Possiamo usare .groupby anche con piú variabili

df_group_two = df[['drive-wheels', 'body-style', 'price']]
group_test1 = df_group_two.groupby(['drive-wheels', 'body-style'], as_index=False).mean()
group_test1
```

```python
#Con questi dati possiamo creare una tabella dinamica metodo .pivot()

grouped_pivot = group_test1.pivot(index='drive-wheels', columns='body-style')
grouped_pivot
```

```python
#Vedo che ci sono dei valori NaN, possiamo scambiarli con 0
grouped_pivot = grouped_pivot.fillna(0)
grouped_pivot
```

```python
#Prezzo medio di ogni macchina in base al 'body-style'

df_group_three = df[['body-style', 'price']]

df_group_three = df_group_three.groupby(['body-style'], as_index=False).mean()

df_group_three
```

```python
#-----------------------------------------------#

#Metodo della correlazione di Pearson
#Correlazione è la misura di dipendenza tra due variabili e la casualitá è la relazione causa effetto tra due variabili

df.corr()
```

```python
#I dati estratti sono di difficile comprensione soprattutto quando la mole acquisisce una dimensione considerevole
#Posso estrarre il valore P mi dice quanto la correlazione tra due variabili sia statisticamente significativa
#Importiamo la libreria SciPy

from scipy import stats
```

```python

#Calcoliamo il coeff di Pearson p per 'wheel-base' vs 'price'

pearson_coef, p_value = stats.pearsonr(df['wheel-base'], df['price'])
print("The Pearson Correlation Coefficient is", pearson_coef, " with a P-value of P =", p_value)  
```

```python
#Calcoliamo il coeff di Pearson per 'horse-power' vs 'price'

pearson_coef, p_value = stats.pearsonr(df['horsepower'], df['price'])
print("The Pearson Correlation Coefficient is", pearson_coef, " with a P-value of P = ", p_value)   
```

```python
#---------------------------------------------------------------#

#Analisi di variazione ANOVA, verifica se ci sono variazioni significative tra i valori medi di due gruppi

#Creo un gruppo di test nuovo con 'drive-wheels' e 'price' e faccio un .groupby()

grouped_test2 = df_group_two[['drive-wheels', 'price']].groupby(['drive-wheels'])
grouped_test2.head(2)
```

```python
df_group_two
```

```python
# Ottengo i valori con il metodo .get_group()

grouped_test2.get_group('4wd')['price']
```

```python
#Applico ANOVA con .f_oneway() per il valore F e P per tutti e 3 i tipi di trazione

f_val, p_val = stats.f_oneway(grouped_test2.get_group('fwd')['price'], grouped_test2.get_group('rwd')['price'], grouped_test2.get_group('4wd')['price'])

print('ANOVA results: F=', f_val, 'P=', p_val)
```

```python
#Applico ANOVA con .f_oneway() per il valore F e P per i tipi di trazione a coppie

#FWD e RWD

f_val, p_val = stats.f_oneway(grouped_test2.get_group('fwd')['price'], grouped_test2.get_group('rwd')['price'])

print('ANOVA results FWV vs RWD: F=', f_val, 'P=', p_val)

#RWD vs 4WD

f_val1, p_val1 = stats.f_oneway(grouped_test2.get_group('rwd')['price'], grouped_test2.get_group('4wd')['price'])

print('ANOVA results RWV vs 4WD: F=', f_val1, 'P=', p_val1)

#FWD vs 4WD

f_val2, p_val2 = stats.f_oneway(grouped_test2.get_group('fwd')['price'], grouped_test2.get_group('4wd')['price'])

print('ANOVA results FWV vs 4WD: F=', f_val2, 'P=', p_val2)

```
